Identify unique column

```python
hotels_id_cols = ['id', 'hotelcode', 'hotelid']
rates_id_cols = ['id', 'refid', 'hotelid']
for h_id in hotels_id_cols:
    for r_id in rates_id_cols:
        num_r_h = rates_df[r_id].nunique()
        common = len( set(hotel_df[h_id]) & set(rates_df[r_id]) )
        print('hotel column used:', h_id, '. rate column used:', r_id, '. Pct common:', (common / num_r_h) )
        print('num_r_h:', num_r_h)
```

Hotel ID has most common values in both the Table

```python
common_id = set(rates_df.columns) & set(hotel_df.columns)
print("Common ID fields:", common_id)
for id_field in common_id:
    if rates_df[id_field].is_unique and hotel_df[id_field].is_unique:
        print(f"Valid ID field for joining: {id_field}")
        break
else:
    print("No valid ID field found for joining.")
   
```
